```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Reading the image from the present directory
image_path = r"G:\My Drive\Project_NNJ_Full_version_july11\Project_NNJ\Morph_Final_Jaundice\Segmented_Jaundice_image_2.png"
image_path1 = r"G:\My Drive\Project_NNJ_Full_version_july11\Project_NNJ\Morph_final_Normal\Morph_normal_image_1.png"
image = cv2.imread(image_path)
image1 = cv2.imread(image_path1)

# Split the image into BGR channels
b, g, r = cv2.split(image)

# Apply CLAHE to each channel
clahe = cv2.createCLAHE(clipLimit=2.0)
b = clahe.apply(b) 
g = clahe.apply(g) 
r = clahe.apply(r)

# Merge the channels back together
final_img = cv2.merge((b, g, r))

b1, g1, r1 = cv2.split(image1)
clahe = cv2.createCLAHE(clipLimit=2.0)
b1 = clahe.apply(b1) 
g1 = clahe.apply(g1) 
r1 = clahe.apply(r1)

# Merge the channels back together
final_img1 = cv2.merge((b1, g1, r1))


# Ordinary thresholding the same image
_, ordinary_img = cv2.threshold(image, 155, 255, cv2.THRESH_BINARY, cv2.THRESH_OTSU)

# Convert the images from BGR to RGB
ordinary_img = cv2.cvtColor(ordinary_img, cv2.COLOR_BGR2RGB)

final_img = cv2.cvtColor(final_img, cv2.COLOR_BGR2RGB)
final_img1 = cv2.cvtColor(final_img1, cv2.COLOR_BGR2RGB)

# Plot both images side by side
plt.figure(figsize=(10,10))
plt.subplot(3,2,1)
plt.imshow(ordinary_img)
plt.title('Ordinary Threshold')
plt.subplot(3,2,2)
plt.imshow(final_img)
plt.title('CLAHE_jaundice_Image')

# Plot the histograms for each color channel before and after CLAHE
bins = 3
plt.subplot(3,2,3)
plt.hist(image[:,:,0].ravel(), bins=bins, color='b', alpha=0.5, label='B')
plt.hist(image[:,:,1].ravel(), bins=bins, color='g', alpha=0.5, label='G')
plt.hist(image[:,:,2].ravel(), bins=bins, color='r', alpha=0.5, label='R')
plt.legend()
plt.title('Histogram before CLAHE')
plt.subplot(3,2,4)
plt.hist(final_img[:,:,0].ravel(), bins=bins, color='b', alpha=0.5, label='B')
plt.hist(final_img[:,:,1].ravel(), bins=bins, color='g', alpha=0.5, label='G')
plt.hist(final_img[:,:,2].ravel(), bins=bins, color='r', alpha=0.5, label='R')
plt.legend()
plt.title('Histogram after CLAHE')
plt.subplot(3,2,5)
plt.imshow(final_img1)
plt.title('CLAHE_normal_Image')
plt.subplot(3,2,6)
plt.hist(final_img1[:,:,0].ravel(), bins=bins, color='b', alpha=0.5, label='B')
plt.hist(final_img1[:,:,1].ravel(), bins=bins, color='g', alpha=0.5, label='G')
plt.hist(final_img1[:,:,2].ravel(), bins=bins, color='r', alpha=0.5, label='R')
plt.title('Normal image Histogram after CLAHE')
plt.subplot(3,2,5)
plt.show()
# view image in 3 channels


```
