```python
import os
import cv2
import numpy as np

def open_image(image):
    kernel = np.ones((15 , 15), np.uint8)
    opening = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    return opening

# specify the input and output directories
input_dir = r'G:\My Drive\Project_NNJ_Full_version_july11\Project_NNJ\Segmented_Jaundice_images'
output_dir = r'G:\My Drive\Project_NNJ_Full_version_july11\Project_NNJ\Morph_Final_Jaundice'

# create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# iterate over all images in the input directory
for filename in os.listdir(input_dir):
    # construct the full path to the input image
    input_path = os.path.join(input_dir, filename)
    
    # read the input image
    im = cv2.imread(input_path)
    
    # perform morphological open operation on the image
    image = open_image(im)
    
    # construct the full path to the output image
    output_path = os.path.join(output_dir, filename)
    
    # save the resulting image to the output directory
    cv2.imwrite(output_path, image)
```
